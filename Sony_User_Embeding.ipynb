```python
import os
import pickle5 as pickle
import numpy as np
from pathlib import Path
import random
import string
import warnings
warnings.filterwarnings('ignore') 
```

```python
def load_pkl(filename):
    filename= filename
    data= None
    with open(filename, "rb") as handle:
        data= pickle.load(handle)
        handle.close()
    return data
```

# Code

```python
#Loading movie information (Either text, audio or viddeo)
text_enc= load_pkl("/content/drive/MyDrive/sony_IITPatna/new_fickscore_meta_processed.obj")

#Loading user information
complete_enc= load_pkl("/content/drive/MyDrive/sony_IITPatna/complete_rating_user_embedding.obj")

print(len(complete_enc))
print(len(text_enc))
```

```python
complete_enc_v2 = []
for i in range(763):
  if len(complete_enc[i]) != 0: # Removing zero length items from complete_enc
    complete_enc_v2.append(complete_enc[i])
```

```python
finalop = []
for j in complete_enc_v2:
  kl = []
  #print(j)
  id = j[0][0]
  #print(id)
  for i in j:
  #l = []
      #print(i)
      if i[2][0] == '1' :
         kl.append(i[1])
  p = {id : kl}          # Making list of dictionaries of movie_id liked (Rating = 1) by the user 
  finalop.append(p)      # Format: [{'user_id':[movie_id]}]
```

```python
finalop[0]
```

```python
user_embed = []
for j in range(len(finalop)):
  id  = list(finalop[j].keys())[0]
  k = finalop[j].get(id)
  lp = []
  for i in k:
     if i in text_enc.keys():
        lp.append(text_enc.get(i))
  lp = np.array(lp)                      # Making list of dictionaries of movie_id liked (Rating = 1) by the user by taking the average      
                                         # Format: [{'user_id':[textual embedding of movie_ids]}]
  pop = np.average(lp, axis=0)
  mk = {id:pop}
  user_embed.append(mk)
```

```python
user_embed[0]
```

```python
dict_user_embed = dict()
for i in range(len(user_embed)):
  #print(i)
  bnb = list(user_embed[i].keys())[0]
  # print(bnb)
  value = user_embed[i][bnb]               
  dict_user_embed[bnb] = value    # Making dictionary of movie_id liked (Rating = 1) by the user  
                                  # Format: {'user_id':[textual embedding of movie_ids]}
```
