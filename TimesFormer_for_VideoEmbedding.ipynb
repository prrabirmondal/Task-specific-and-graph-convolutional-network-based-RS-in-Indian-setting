# Training and Validate Function

```python
import torch.nn as nn
model = TimeSformer(img_size=224, num_classes=600, num_frames=16, attention_type='divided_space_time',  pretrained_model=str(model_file))
for param in model.parameters():
    param.requires_grad = False

model.model.head = nn.Sequential(nn.Linear(768,512),
                 nn.LayerNorm(512),
                 nn.Linear(512,18))


# ## Engine 
# training function
def train(model, dataloader, optimizer, criterion, train_data, device):
    print('Training')
    model.train()
    counter = 0
    train_running_loss = 0.0
    for i, data in tqdm(enumerate(dataloader), total=int(len(train_data)/dataloader.batch_size)):
        counter += 1
        videos, labels = data
        data = videos.to(device)
        target = labels.to(device)
        #print(data.shape,target.shape)
        #data, target = data['image'].to(device), data['label'].to(device)
        optimizer.zero_grad()
        outputs = model(data)
        # apply sigmoid activation to get all the outputs between 0 and 1
        outputs = torch.sigmoid(outputs)
        #print("Printing.....")
        #print(outputs)
        #print(type(outputs) , outputs.shape)
        #print("Printing.....Target")
        #print(target)
        #print(type(target) , target.shape)        
        loss = criterion(outputs, target.to(torch.float32))
        train_running_loss += loss.item()
        # backpropagation
        loss.backward()
        # update optimizer parameters
        optimizer.step()
        scheduler.step()
        #print(optimizer.param_groups[0]["lr"])
	
        
    train_loss = train_running_loss / counter
    return train_loss

# validation function
def validate(model, dataloader, criterion, val_data, device):
    print('Validating')
    model.eval()
    counter = 0
    val_running_loss = 0.0
    running_accuracy = 0.0
    running_accuracy_class = torch.zeros(18).to(device)
    with torch.no_grad():
        for i, data in tqdm(enumerate(dataloader), total=int(len(val_data)/dataloader.batch_size)):
            counter += 1
            videos, labels = data
            data = videos.to(device)
            target = labels.to(device)
            #data, target = data['image'].to(device), data['label'].to(device)
            outputs = model(data)
            # apply sigmoid activation to get all the outputs between 0 and 1
            outputs = torch.sigmoid(outputs)            
            loss = criterion(outputs, target.to(torch.float32))
            val_running_loss += loss.item()

            #Calculating Accuracy        
                   
            target = target.to(torch.float32)
            #print("Outputs:" , outputs )
            #print("Target:" , target )
            
            running_accuracy += multilabel_accuracy(outputs, target, num_labels=18)
            running_accuracy_class += multilabel_accuracy(outputs, target, num_labels=18, average=None)
        
        val_loss = val_running_loss / counter
        accuracy = running_accuracy/counter
        classwise_accuracy = running_accuracy_class / counter
        return val_loss , accuracy ,classwise_accuracy
```

```python
# learning parameters
model = model.to(device)
lr = 0.1
epochs = 40
optimizer = optim.Adam(model.parameters(), lr=lr)
criterion = nn.BCELoss()
scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=0.1, steps_per_epoch=len(train_loader), epochs=40)
print("Model Transfered")

# save the trained model to disk
def save(model):
    torch.save(model, '/content/drive/MyDrive/ml-20m/ml-20m/16_model_movielens_CE.pth')

# start the training and validation
train_loss = []
valid_loss = []
min_valid_loss = 1000.0
for epoch in range(epochs):
    print(f"Epoch {epoch+1} of {epochs}")
    train_epoch_loss = train(
        model, train_loader, optimizer, criterion, train_set, device
    )
    valid_epoch_loss , accuracy_score_epoch , classwise_accuracy_epoch = validate(
        model, test_loader, criterion, test_set, device
    )
    if valid_epoch_loss < min_valid_loss:
        min_valid_loss = valid_epoch_loss
        print("Saving Model........" , min_valid_loss)
        save(model)

    train_loss.append(train_epoch_loss)
    valid_loss.append(valid_epoch_loss)
    print(f"Train Loss: {train_epoch_loss:.4f}")
    print(f'Val Loss: {valid_epoch_loss:.4f}')
    print(f'Val Acc: {accuracy_score_epoch:.4f}')
    print("Classs Wise Acc: ", classwise_accuracy_epoch)
```
